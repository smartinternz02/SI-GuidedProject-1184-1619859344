# Seaborn visualization on Iris Dataset

```python
from seaborn import*
import seaborn as sb
from matplotlib.pyplot import*
import matplotlib.pyplot as mp
```

```python
load_dataset('iris')
```

```python
figure(figsize = (10,5))
subplot(1,2,1)
scatterplot(x="sepal_length",y="sepal_width",data=iris)
subplot(1,2,2)
scatterplot(x="petal_length",y="petal_width",data=iris)
```

```python
figure(figsize=(10,5))
subplot(1,2,1)
lineplot(x="sepal_length",y="sepal_width",data=iris)
subplot(1,2,2)
lineplot(x="petal_length",y="petal_width",data=iris)
```

```python
figure(figsize=(10,5))
subplot(2,2,1)
distplot(iris["sepal_length"])
subplot(2,2,2)
distplot(iris["sepal_width"])
subplot(2,2,3)
distplot(iris["petal_length"])
subplot(2,2,4)
distplot(iris["petal_width"])
```

```python
figure(figsize=(15,5))
subplot(1,2,1)
barplot(x="sepal_length",y="sepal_width",data=iris.loc[iris["species"]=="setosa"])
subplot(1,2,2)
barplot(x="petal_length",y="petal_width",data=iris.loc[iris["species"]=="setosa"])
```

```python
figure(figsize=(15,5))
subplot(1,2,1)
sb.boxplot("sepal_length","sepal_width",data=iris.loc[iris["species"]=="virginica"])
subplot(1,2,2)
sb.boxplot("petal_length","petal_width",data=iris.loc[iris["species"]=="virginica"])
```

```python
catplot("species","sepal_length",data=iris)
catplot("species","petal_width",data=iris)
```
